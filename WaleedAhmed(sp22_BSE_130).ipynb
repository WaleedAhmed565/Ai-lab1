```python
import numpy as np
import matplotlib.pyplot as plt

# Temperature data for each city
temps_data = {
    "Islamabad": [7, 9, 15, 22, 28, 33, 31, 30, 27, 21, 15, 9],
    "Lahore": [10, 12, 18, 26, 33, 37, 36, 35, 31, 25, 19, 12],
    "Karachi": [18, 19, 22, 26, 29, 32, 31, 31, 29, 27, 23, 20],
    "Quetta": [5, 7, 12, 18, 24, 28, 27, 26, 23, 17, 11, 6]
}

temps_array = np.array([temps_data["Islamabad"], temps_data["Lahore"], temps_data["Karachi"], temps_data["Quetta"]])
monthly_avg_temp = np.mean(temps_array, axis=0)

# Generate line plots for each city's temperatures over the year
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
plt.figure(figsize=(14, 7))

# Line plot for each city
for city, temps in temps_data.items():
    plt.plot(months, temps, marker='o', label=city)

plt.title("Monthly Temperature Trends for Major Cities in Pakistan")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(12, 6))
plt.bar(months, monthly_avg_temp, color='skyblue')
plt.title("Average Monthly Temperature Across All Cities")
plt.xlabel("Month")
plt.ylabel("Average Temperature (C)")
plt.grid(axis='y')
plt.show()
print("Monthly Average Temperatures Across All Cities:")
print(monthly_avg_temp)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the hill shape function with a peak at x=5 and height=20
def hill_shape(x):
    return -((x - 5) ** 2) + 20

# Generate x values for plotting
x_values = np.linspace(0, 10, 100)
y_values = hill_shape(x_values)  # Calculate corresponding y values

# Peak position and height found from the simulation
peak_position = 5  # x value at peak
peak_height = hill_shape(peak_position)  # height at the peak

# Plot the hill shape
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label="Hill Shape", color="b")

# Mark the peak position found in the previous code
plt.plot(peak_position, peak_height, 'ro', label=f"Peak at x={peak_position:.2f}, height={peak_height:.2f}")

# Add labels and title
plt.xlabel("Position (x)")
plt.ylabel("Height")
plt.title("Hill Climb Simulation: Hill Shape with Peak")
plt.legend()
plt.grid(True)
plt.show()
```
