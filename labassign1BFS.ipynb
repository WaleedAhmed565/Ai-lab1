```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Define the edges between cities
edges = [
    ("Islamabad", "Rawalpindi"),
    ("Islamabad", "Lahore"),
    ("Islamabad", "Peshawar"),
    ("Rawalpindi", "Peshawar"),
    ("Rawalpindi", "Quetta"),
    ("Peshawar", "Quetta"),
    ("Lahore", "Multan"),
    ("Lahore", "Quetta"),
    ("Multan", "Quetta"),
    ("Multan", "Karachi"),
    ("Quetta", "Karachi")
]
G.add_edges_from(edges)

def bfs_shortest_path(graph, start_city, end_city):
    """
    Finds the shortest path between two cities in an undirected graph using BFS.

    Parameters:
    graph (nx.Graph): The networkx graph object.
    start_city (str): The starting city.
    end_city (str): The target city.

    Returns:
    list: A list of cities representing the shortest path.
    """
    return nx.shortest_path(graph, start_city, end_city)

# Set start and end cities
start = "Islamabad"
end = "Karachi"

# Get the shortest path using BFS
shortest_path = bfs_shortest_path(G, start, end)

# Precompute the layout of the graph for better performance
pos = nx.spring_layout(G)

# Create the plot
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=3000,
        font_size=10, font_weight="bold", edge_color="gray")

# Highlight the edges along the shortest path
path_edges = [(shortest_path[i], shortest_path[i + 1]) for i in range(len(shortest_path) - 1)]
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=3)

# Add a title with the shortest path
plt.title(f"Shortest path from {start} to {end}: {' -> '.join(shortest_path)}", fontsize=12)

# Show the plot
plt.show()

# Print the shortest path
print(f"Shortest path from {start} to {end}: {' -> '.join(shortest_path)}")
```
